Q1. Write a Python function that takes in two arrays of data and calculates the F-value for a variance ratio
test. The function should return the F-value and the corresponding p-value for the test.

```python
Certainly! Here's a Python function using the `scipy` library to calculate the F-value and corresponding p-value for a variance ratio test:

from scipy.stats import f

def calculate_f_test(data1, data2):
    # Calculate the variances of the two datasets
    var1 = np.var(data1, ddof=1)  # ddof=1 for sample variance
    var2 = np.var(data2, ddof=1)
    
    # Calculate the F-value and p-value
    f_value = var1 / var2 if var1 > var2 else var2 / var1  # Variance ratio
    df1 = len(data1) - 1
    df2 = len(data2) - 1
    p_value = f.sf(f_value, df1, df2)  # Survival function (1 - CDF) at F-value

    return f_value, p_value

This function takes two arrays of data (`data1` and `data2`) and computes the F-value and the corresponding p-value for the variance ratio test using the `scipy.stats.f` module.

You can use this function by passing your datasets `data1` and `data2` as arguments.
```

Q2. Given a significance level of 0.05 and the degrees of freedom for the numerator and denominator of an
F-distribution, write a Python function that returns the critical F-value for a two-tailed test.

```python
Certainly! Here's a Python function that calculates the critical F-value for a two-tailed test given a significance level and degrees of freedom for the numerator and denominator:

from scipy.stats import f

def critical_f_value(alpha, dfn, dfd):
    return f.ppf(1 - alpha / 2, dfn, dfd)

This function utilizes the `scipy.stats.f.ppf` function to find the critical F-value corresponding to the given significance level (`alpha`), degrees of freedom for the numerator (`dfn`), and degrees of freedom for the denominator (`dfd`). The `ppf` function gives the value from the F-distribution's percent point function at the given probability.

You can call this function with your desired significance level (`alpha`) and degrees of freedom values (`dfn` and `dfd`) to obtain the critical F-value for a two-tailed test.
```

Q4.The variances of two populations are known to be 10 and 15. A sample of 12 observations is taken from
each population. Conduct an F-test at the 5% significance level to determine if the variances are
significantly different.

```python
To conduct an F-test for comparing the variances of two populations, let's calculate the F-value and determine if the variances are significantly different.

The F-value can be calculated using the formula:

\[ F = \frac{s_1^2}{s_2^2} \]

Where:
- \( s_1^2 \) and \( s_2^2 \) are the sample variances of the two populations.

Given:
- Sample variance 1 (\( s_1^2 \)) = 10
- Sample variance 2 (\( s_2^2 \)) = 15
- Sample size for both populations = 12

Let's compute the F-value and compare it to the critical F-value at the 5% significance level for degrees of freedom (dfn = 11, dfd = 11) to determine if the variances are significantly different.

from scipy.stats import f

# Given data
s1_squared = 10  # Variance of population 1
s2_squared = 15  # Variance of population 2
n = 12  # Sample size for both populations

# Calculating the F-value
F_value = s1_squared / s2_squared

# Degrees of freedom for numerator and denominator
dfn = n - 1
dfd = n - 1

# Critical F-value at 5% significance level for dfn and dfd
alpha = 0.05
critical_F_value = f.ppf(1 - alpha / 2, dfn, dfd)

# Outputting the F-value and the critical F-value
print(f"F-value: {F_value}")
print(f"Critical F-value: {critical_F_value}")

# Checking if the variances are significantly different
if F_value > critical_F_value or 1 / F_value > critical_F_value:
    print("The variances are significantly different at the 5% significance level.")
else:
    print("The variances are not significantly different at the 5% significance level.")

This code calculates the F-value based on the sample variances and compares it to the critical F-value at the 5% significance level for the given degrees of freedom. If the calculated F-value is greater than the critical F-value or its reciprocal is greater than the critical F-value, then the variances are significantly different at the 5% significance level.
```

Q5. A manufacturer claims that the variance of the diameter of a certain product is 0.005. A sample of 25
products is taken, and the sample variance is found to be 0.006. Conduct an F-test at the 1% significance
level to determine if the claim is justified.

```python
To test the manufacturer's claim about the variance of the product's diameter, we'll conduct an F-test at a 1% significance level using the sample variance.

The F-test compares the sample variance to the claimed population variance. If the calculated F-value exceeds the critical F-value for the degrees of freedom corresponding to the sample, it suggests a significant difference between the sample variance and the claimed population variance.

Given:
- Claimed population variance = 0.005
- Sample variance = 0.006
- Sample size = 25

Let's perform the F-test:

from scipy.stats import f

# Given data
claimed_variance = 0.005  # Claimed population variance
sample_variance = 0.006  # Sample variance
sample_size = 25  # Sample size

# Calculating the F-value
F_value = sample_variance / claimed_variance

# Degrees of freedom for numerator and denominator
dfn = sample_size - 1
dfd = sample_size - 1

# Critical F-value at 1% significance level for dfn and dfd
alpha = 0.01
critical_F_value = f.ppf(1 - alpha, dfn, dfd)

# Outputting the F-value and the critical F-value
print(f"F-value: {F_value}")
print(f"Critical F-value: {critical_F_value}")

# Checking if the claim is justified or not
if F_value < critical_F_value:
    print("The claim about the population variance is justified at the 1% significance level.")
else:
    print("The claim about the population variance is not justified at the 1% significance level.")

This code calculates the F-value using the sample variance and the claimed population variance, compares it to the critical F-value at the 1% significance level, and determines if the claim about the population variance is justified or not.
```

Q6. Write a Python function that takes in the degrees of freedom for the numerator and denominator of an
F-distribution and calculates the mean and variance of the distribution. The function should return the
mean and variance as a tuple.

```python
Certainly! Here's a Python function that computes the mean and variance of an F-distribution given the degrees of freedom for the numerator and denominator:

from scipy.stats import f

def f_distribution_mean_var(dfn, dfd):
    # Calculate mean and variance of the F-distribution
    mean = dfd / (dfd - 2) if dfd > 2 else None
    variance = (2 * dfd**2 * (dfn + dfd - 2)) / (dfn * (dfd - 2)**2 * (dfd - 4)) if dfd > 4 else None
    return mean, variance

# Example usage
degrees_of_freedom_numerator = 5
degrees_of_freedom_denominator = 10

mean, variance = f_distribution_mean_var(degrees_of_freedom_numerator, degrees_of_freedom_denominator)
print(f"Mean of F-distribution: {mean}")
print(f"Variance of F-distribution: {variance}")

This function, `f_distribution_mean_var`, takes the degrees of freedom for the numerator (`dfn`) and denominator (`dfd`) as inputs and computes the mean and variance of the F-distribution. The computed mean and variance are returned as a tuple. Adjust the input parameters `degrees_of_freedom_numerator` and `degrees_of_freedom_denominator` to match your desired degrees of freedom for testing.
```

Q7. A random sample of 10 measurements is taken from a normal population with unknown variance. The
sample variance is found to be 25. Another random sample of 15 measurements is taken from another
normal population with unknown variance, and the sample variance is found to be 20. Conduct an F-test
at the 10% significance level to determine if the variances are significantly different.

```python
To perform an F-test for comparing the variances of two populations, we can use Python with Scipy's `f` distribution from the `scipy.stats` module. Here's how you can conduct the F-test:

from scipy.stats import f

# Given sample variances
variance_1 = 25
variance_2 = 20

# Degrees of freedom for the samples
n1 = 10
n2 = 15

# Calculating the F-value
F = variance_1 / variance_2

# Degrees of freedom for the F-distribution
dfn = n1 - 1
dfd = n2 - 1

# Calculating the p-value
p_value = 2 * min(f.cdf(F, dfn, dfd), 1 - f.cdf(F, dfn, dfd))

alpha = 0.10  # Significance level

# Performing the F-test
if p_value < alpha:
    print(f"Reject null hypothesis. The variances are significantly different.")
else:
    print(f"Fail to reject null hypothesis. There is no significant difference in variances.")

This code computes the F-value using the ratio of variances and then uses Scipy's `f` distribution to calculate the p-value for the F-test. Adjust the `alpha` value for your desired significance level (10% in this case). If the p-value is less than the chosen significance level, you can reject the null hypothesis, indicating that the variances are significantly different.
```

Q8. The following data represent the waiting times in minutes at two different restaurants on a Saturday
night: Restaurant A: 24, 25, 28, 23, 22, 20, 27; Restaurant B: 31, 33, 35, 30, 32, 36. Conduct an F-test at the 5%
significance level to determine if the variances are significantly different.

```python
Certainly! To conduct an F-test in Python to compare the variances of the waiting times at two different restaurants, you can use the following code:

from scipy.stats import f

# Waiting times data for two restaurants
restaurant_A = [24, 25, 28, 23, 22, 20, 27]
restaurant_B = [31, 33, 35, 30, 32, 36]

# Calculate variances
variance_A = np.var(restaurant_A, ddof=1)
variance_B = np.var(restaurant_B, ddof=1)

# Degrees of freedom
n_A = len(restaurant_A)
n_B = len(restaurant_B)
dfn = n_A - 1
dfd = n_B - 1

# Calculating the F-value
F = variance_A / variance_B

# Calculating the p-value
p_value = 2 * min(f.cdf(F, dfn, dfd), 1 - f.cdf(F, dfn, dfd))

alpha = 0.05  # Significance level

# Performing the F-test
if p_value < alpha:
    print("Reject null hypothesis. The variances are significantly different.")
else:
    print("Fail to reject null hypothesis. There is no significant difference in variances.")

This code first calculates the variances for the waiting times at each restaurant, then computes the F-value and p-value using the `scipy.stats` module. Finally, it performs the F-test at a 5% significance level. If the p-value is less than 0.05, it indicates that the variances are significantly different between the two restaurants.
```

Q9. The following data represent the test scores of two groups of students: Group A: 80, 85, 90, 92, 87, 83;
Group B: 75, 78, 82, 79, 81, 84. Conduct an F-test at the 1% significance level to determine if the variances
are significantly different.

```python
Certainly! Here's how you can conduct an F-test in Python to compare the variances of the test scores between two groups of students:

from scipy.stats import f

# Test scores data for two groups
group_A = [80, 85, 90, 92, 87, 83]
group_B = [75, 78, 82, 79, 81, 84]

# Calculate variances
variance_A = np.var(group_A, ddof=1)
variance_B = np.var(group_B, ddof=1)

# Degrees of freedom
n_A = len(group_A)
n_B = len(group_B)
dfn = n_A - 1
dfd = n_B - 1

# Calculating the F-value
F = variance_A / variance_B

# Calculating the p-value
p_value = 2 * min(f.cdf(F, dfn, dfd), 1 - f.cdf(F, dfn, dfd))

alpha = 0.01  # Significance level

# Performing the F-test
if p_value < alpha:
    print("Reject null hypothesis. The variances are significantly different.")
else:
    print("Fail to reject null hypothesis. There is no significant difference in variances.")

This Python code uses the `scipy.stats` module to compute the F-value and p-value for the F-test. If the p-value is less than 0.01, it indicates that the variances are significantly different between Group A and Group B. Otherwise, if the p-value is greater than 0.01, there's no significant difference in variances.
```
