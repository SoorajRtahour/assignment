Q1. What is MongoDB? Explain non-relational database in short. In which scenarios it is preferred to use MongoDB over SQL database?

```python
MongoDB is a popular open-source, document-oriented NoSQL database management system. It stores data in a flexible, JSON-like format known as BSON (Binary JSON). MongoDB's design focuses on scalability, performance, and ease of development, making it suitable for a wide range of applications.

A non-relational database, also known as a NoSQL database, is a type of database system that does not use the traditional tabular structure of relational databases like SQL. Instead, NoSQL databases use various data models, such as key-value, document, columnar, or graph, to store and manage data. This approach allows for more flexible and efficient handling of data that doesn't fit neatly into a rigid table-based schema.

Scenarios where MongoDB is preferred over SQL databases:

1. Flexible Schema: In MongoDB, each document can have a different structure, allowing you to store data without requiring a predefined schema. This is particularly useful when dealing with dynamic or evolving data.

2. Semi-Structured and Unstructured Data: MongoDB excels in storing semi-structured and unstructured data, such as JSON-like documents. It's great for scenarios where data can vary in structure.

3. Scalability: MongoDB's distributed architecture makes it well-suited for horizontal scaling, allowing you to handle large amounts of data and high traffic loads across multiple servers.

4. Agile Development: MongoDB's schema-less nature and support for dynamic changes make it suitable for agile development environments where requirements might change frequently.

5. Real-time Applications: MongoDB's ability to store and retrieve data quickly makes it suitable for real-time applications, such as social media feeds, IoT devices, and gaming leaderboards.

6. Big Data and Analytics: When dealing with large volumes of data, MongoDB's ability to shard and distribute data efficiently can provide performance benefits.

7. Document-Oriented Data: If your application's data is naturally organized into documents (e.g., blog posts, user profiles), MongoDB's document-based approach simplifies data storage and retrieval.

8. Developer Productivity: Developers who are comfortable with JavaScript or similar languages might find MongoDB's query language more approachable than SQL.

9. Geospatial Data: MongoDB provides robust support for geospatial indexing and querying, making it suitable for applications that involve location-based data.

It's important to note that the choice between MongoDB and SQL databases depends on the specific requirements and nature of your application. Each has its own strengths and weaknesses, and the decision should be based on factors such as data structure, scalability needs, query complexity, and developer familiarity.
```

Q2. State and Explain the features of MongoDB.

```python
MongoDB offers a variety of features that make it a popular choice for many applications. Here are some key features of MongoDB:

01. Document-Oriented: MongoDB stores data in flexible, JSON-like documents called BSON (Binary JSON). Each document can have a different structure, allowing for easy representation of complex, hierarchical data.

02. Flexible Schema: MongoDB's schema-less design enables you to store data without a predefined structure. This is especially beneficial when dealing with evolving or dynamic data.

03. Scalability: MongoDB supports horizontal scaling through sharding, allowing you to distribute data across multiple servers to handle increased workloads. This enables seamless expansion as your application grows.

04. High Availability: MongoDB supports replica sets, which are groups of MongoDB servers that maintain redundant copies of data. If one server fails, another can take over, ensuring data availability and minimizing downtime.

05. Indexing: MongoDB provides a wide range of indexing options to optimize query performance. This includes support for single-field and compound indexes, as well as text and geospatial indexes.

06. Aggregation Framework: MongoDB offers a powerful aggregation framework that allows you to perform complex data transformations and calculations directly within the database. This is particularly useful for analytics and reporting.

07. Geospatial Capabilities: MongoDB includes support for geospatial data, allowing you to store and query location-based information efficiently. This is valuable for applications involving maps, GPS data, and location-based services.

08. Ad Hoc Queries: MongoDB's query language enables developers to perform ad hoc queries against the data. The query language is expressive and resembles JavaScript, making it user-friendly for developers.

09. JSON-Like Query Language: MongoDB's query language is based on JSON syntax, which is familiar to many developers and is intuitive for creating complex queries.

10. Automatic Sharding: MongoDB's sharding feature automatically distributes data across multiple shards, allowing for efficient distribution and management of data as your application scales.

11. Automatic Balancing: The MongoDB balancer monitors the distribution of data across shards and automatically moves data to balance the load, ensuring even data distribution.

12. Rich Ecosystem: MongoDB has a strong and active community, offering a wide range of tools, libraries, and resources to support developers. This includes official drivers for various programming languages.

13. Change Streams: MongoDB provides change streams, allowing applications to subscribe to real-time notifications about changes in the data. This is useful for implementing features like real-time updates and triggers.

14. Full-Text Search: MongoDB offers a full-text search capability, allowing you to search and index text content within documents.

15. Security Features: MongoDB provides features such as access control, authentication mechanisms, encryption at rest, and field-level access control to help secure your data.

These features collectively make MongoDB a versatile choice for various applications, including content management systems, e-commerce platforms, real-time analytics, mobile apps, IoT applications, and more. However, it's important to carefully consider your application's requirements before choosing any database technology.
```

Q3. Write code to connect MongoDB to python. Also, create a database and a collection in MongoDB.

```python
Sure, here's an example code snippet that demonstrates how to connect to MongoDB using Python using the pymongo library, create a database, and create a collection within that database:

python

from pymongo import MongoClient

# Connect to the MongoDB server running on localhost
client = MongoClient('localhost', 27017)

# Create or access a database named 'mydatabase'
database = client['mydatabase']

# Create or access a collection named 'mycollection' within the 'mydatabase'
collection = database['mycollection']

# Create a sample document to insert into the collection
sample_document = {
    'name': 'John Doe',
    'age': 30,
    'email': 'johndoe@example.com'
}

# Insert the document into the collection
insert_result = collection.insert_one(sample_document)

# Print the inserted document's ID
print(f"Inserted document ID: {insert_result.inserted_id}")

# Close the connection
client.close()
In this code, we're using the pymongo library to connect to the MongoDB server running on localhost at the default port (27017). We create a database named 'mydatabase' and a collection named 'mycollection' within that database. We then create a sample document and insert it into the collection using the insert_one() method. Finally, we print the ID of the inserted document and close the connection to the MongoDB server.

Before running this code, make sure you have the pymongo library installed. You can install it using the following command:

pip install pymongo
Also, ensure that your MongoDB server is up and running on localhost at port 27017. If your MongoDB server is running on a different host or port, modify the connection parameters accordingly.
```

Q4. Using the database and collection created in question number 3, write a code to insert one record, and insert many records. Use the find() and find_one() methods to print the inserted record.

```python
Certainly! Here's how you can insert one record and insert multiple records into the previously created collection, and then use the find() and find_one() methods to retrieve and print the inserted records:

python

from pymongo import MongoClient

# Connect to the MongoDB server running on localhost
client = MongoClient('localhost', 27017)

# Access the 'mydatabase' database and 'mycollection' collection
database = client['mydatabase']
collection = database['mycollection']

# Insert one record
one_record = {
    'name': 'Jane Smith',
    'age': 25,
    'email': 'janesmith@example.com'
}
insert_one_result = collection.insert_one(one_record)
print("Inserted one record with ID:", insert_one_result.inserted_id)

# Insert many records
many_records = [
    {'name': 'Alice Johnson', 'age': 28, 'email': 'alice@example.com'},
    {'name': 'Bob Williams', 'age': 32, 'email': 'bob@example.com'},
    {'name': 'Eve Davis', 'age': 22, 'email': 'eve@example.com'}
]
insert_many_result = collection.insert_many(many_records)
print("Inserted many records with IDs:", insert_many_result.inserted_ids)

# Find and print the inserted records
print("\nInserted records:")
for record in collection.find():
    print(record)

# Find and print one inserted record
print("\nFind one record:")
one_record_found = collection.find_one({'name': 'Jane Smith'})
print(one_record_found)

# Close the connection
client.close()
In this code, we first insert one record using the insert_one() method and print the ID of the inserted record. Then, we insert multiple records using the insert_many() method and print the IDs of the inserted records.

After insertion, we use the find() method to retrieve and print all records in the collection. We also use the find_one() method to retrieve and print a specific record based on the query criteria.

Remember to modify the connection parameters if your MongoDB server is running on a different host or port.
```

Q5. Explain how you can use the find() mwthod to query the MongoDB database. Write a simple code to demonstrate this. 

```python
The find() method in MongoDB is used to query a collection and retrieve documents that match a specific query criteria. This method returns a cursor, which you can iterate over to access the retrieved documents. The query criteria are specified using a dictionary-based query document.

Here's a breakdown of how to use the find() method and a code example to demonstrate this:

1. Syntax:

scss

cursor = collection.find(query, projection)
query: A dictionary specifying the query criteria. This dictionary contains fields and their corresponding values to filter the documents.
projection (optional): A dictionary that specifies which fields to include or exclude in the returned documents. It's used to control the fields' visibility in the query results.

2. Code Example:

Let's say we want to retrieve all documents from the 'mycollection' collection where the 'age' field is greater than or equal to 30. We also want to exclude the 'email' field from the query results.

python

from pymongo import MongoClient

# Connect to the MongoDB server running on localhost
client = MongoClient('localhost', 27017)

# Access the 'mydatabase' database and 'mycollection' collection
database = client['mydatabase']
collection = database['mycollection']

# Query criteria: age >= 30
query = {'age': {'$gte': 30}}

# Projection: Exclude 'email' field
projection = {'email': 0}

# Find documents that match the query
cursor = collection.find(query, projection)

# Iterate over the cursor and print the retrieved documents
print("Documents matching the query:")
for document in cursor:
    print(document)

# Close the connection
client.close()

In this code, we use the find() method with a query to retrieve documents from the collection where the 'age' field is greater than or equal to 30. We also use the projection parameter to exclude the 'email' field from the query results. The retrieved documents are then iterated over and printed.

Remember to adjust the connection parameters as needed based on your MongoDB setup.
```

Q6. Explain the sort() method. Give an example to demonstrate sorting in MongoDB.

```python
The sort() method in MongoDB is used to specify how the query results should be sorted. It allows you to specify one or more fields by which the documents in the query result should be ordered. You can sort the documents in ascending (1) or descending (-1) order based on the specified field(s).

Here's how the sort() method works and an example to demonstrate sorting in MongoDB:

Syntax:

python

cursor = collection.find(query).sort(sort_key, sort_order)
query: The query criteria used to filter the documents.
sort_key: The field by which the documents should be sorted.
sort_order: The sorting order. Use 1 for ascending and -1 for descending.
Example:

Suppose we want to retrieve documents from the 'mycollection' collection and sort them in descending order based on the 'age' field.

python

from pymongo import MongoClient

# Connect to the MongoDB server running on localhost
client = MongoClient('localhost', 27017)

# Access the 'mydatabase' database and 'mycollection' collection
database = client['mydatabase']
collection = database['mycollection']

# Query criteria: Retrieve all documents
query = {}

# Sort by 'age' field in descending order
sort_key = 'age'
sort_order = -1

# Find documents, sort them, and retrieve the cursor
cursor = collection.find(query).sort(sort_key, sort_order)

# Iterate over the cursor and print the sorted documents
print("Sorted documents:")
for document in cursor:
    print(document)

# Close the connection
client.close()

In this example, we use the sort() method to retrieve documents from the collection and sort them in descending order based on the 'age' field. The sort_key is set to 'age', and the sort_order is set to -1. The retrieved documents are then iterated over and printed in the specified sorting order.

Remember to adapt the connection parameters based on your MongoDB configuration.
```

Q7. Explian why delete_one(), delete_many(), and drop() is used.

```python
In MongoDB, the delete_one(), delete_many(), and drop() methods are used to remove documents or collections from the database. Each of these methods serves a different purpose:

1. delete_one() Method:
The delete_one() method is used to remove a single document that matches a specified query from a collection.

Syntax:

python

result = collection.delete_one(query)
query: The query criteria to identify the document to be deleted.
The delete_one() method returns a DeleteResult object that provides information about the deletion operation, including the number of documents deleted (which will be either 0 or 1).

2. delete_many() Method:
The delete_many() method is used to remove multiple documents that match a specified query from a collection.

Syntax:

python

result = collection.delete_many(query)
query: The query criteria to identify the documents to be deleted.
The delete_many() method also returns a DeleteResult object that contains information about the deletion operation, including the number of documents deleted.

3. drop() Method:
The drop() method is used to remove an entire collection from the database. It essentially deletes the collection itself and all of its documents.

Syntax:

python

collection.drop()
The drop() method does not take any arguments. It directly removes the entire collection specified.

These methods are used to manage data within a MongoDB database:

delete_one() is useful when you want to remove a specific document from a collection.
delete_many() is suitable when you want to delete multiple documents that match a certain query.
drop() is used when you want to completely remove a collection, including all its documents.

It's important to use these methods carefully, as deletions are irreversible. Always double-check your queries and understand the implications before executing them.
```
