Q1. What is an Exception in python? Write the difference between Exception and Syntax errors.

```python
In Python, an exception is an unexpected event that occurs during the execution of a program and disrupts its normal flow. When an exceptional situation arises, such as division by zero, accessing an index beyond the bounds of a list, or attempting to convert an invalid data type, Python raises an exception to indicate that something went wrong. Exceptions can occur at runtime and can be caused by various factors like user input, external resources, or programming errors.

There are different types of exceptions in Python, and they are represented by specific classes. Some common built-in exception types in Python include ValueError, TypeError, ZeroDivisionError, IndexError, FileNotFoundError, etc. Additionally, Python allows you to create your own custom exception classes to handle specific situations unique to your program.

Now, let's discuss the difference between exceptions and syntax errors:

Exceptions:

Exceptions occur during the runtime of a Python program.
They are raised when an error or unexpected situation occurs while the code is executing.
Examples of exceptions include ZeroDivisionError, ValueError, TypeError, etc.
Exceptions can often be handled using try and except blocks to prevent the program from crashing and to provide graceful error handling.
Syntax Errors:

Syntax errors occur during the parsing phase, which is the initial stage when Python is trying to understand the code structure.
They are raised when the code violates the rules of the Python language grammar, making it unable to understand the code.
Examples of syntax errors include missing colons, unmatched parentheses, invalid variable names, etc.
Syntax errors need to be fixed in the code before the program can run. The code won't execute until the syntax errors are resolved.
Here's an example to illustrate the difference between an exception and a syntax error:

Exception Example:

python

try:
    num1 = int(input("Enter a number: "))
    result = 10 / num1  # This line may raise a ZeroDivisionError if num1 is 0.
    print("Result:", result)
except ZeroDivisionError:
    print("Cannot divide by zero.")
In this example, if the user enters 0 as the num1, a ZeroDivisionError exception will be raised and handled by the except block.

Syntax Error Example:

python

# This code has a syntax error.
for i in range(5)
    print(i)
In this example, there's a missing colon (:) after range(5), which will cause a syntax error. The code won't execute until the colon is added to the end of the for statement.
```

Q2. What happens when an exception is not handled? Explain with an example.

```python
When an exception is not handled in Python, it will cause the program to terminate abruptly, and an error message (called a traceback) will be displayed, indicating the type of exception and the line number where it occurred. This behavior can be problematic for the user and is generally undesirable in a production environment, as it leads to an ungraceful termination of the program.

Let's look at an example to illustrate what happens when an exception is not handled:

python

def divide_numbers(a, b):
    result = a / b
    return result

num1 = 10
num2 = 0

result = divide_numbers(num1, num2)
print("Result:", result)
In this example, we have a function divide_numbers that takes two arguments and performs division. We call this function with num1 = 10 and num2 = 0. Since dividing by zero is not allowed in mathematics and Python raises a ZeroDivisionError in this situation, we will encounter an unhandled exception.

When we run this code, the following output will be displayed:

arduino

Traceback (most recent call last):
  File "example.py", line 7, in <module>
    result = divide_numbers(num1, num2)
  File "example.py", line 2, in divide_numbers
    result = a / b
ZeroDivisionError: division by zero
The traceback shows the type of exception (ZeroDivisionError) and the line numbers where the exception was raised (File "example.py", line 2). It clearly indicates that a division by zero occurred during the execution of the divide_numbers function.

Since we did not handle the ZeroDivisionError with a try-except block, the program terminates immediately after displaying the traceback. Any code that was supposed to execute after the point of the exception will not be executed.

To avoid unhandled exceptions, it is a good practice to use try and except blocks to gracefully handle exceptions and provide appropriate error messages or corrective actions when exceptional situations occur during the program execution. This ensures that the program can recover from errors and continue running smoothly, even if unexpected scenarios arise.
```

Q3. Which python statements are used to catch and handle exceptions? Explain with an example.

```python
In Python, the try and except statements are used to catch and handle exceptions. The try block contains the code that may raise an exception, and the except block specifies the code that should be executed to handle the exception if it occurs.

The general syntax for using try and except is as follows:

python

try:
    # Code that may raise an exception
    # ...
except SomeExceptionType:
    # Code to handle the exception
    # ...
Here's an example to illustrate how try and except statements are used to catch and handle exceptions:

python

def divide_numbers(a, b):
    try:
        result = a / b
        return result
    except ZeroDivisionError:
        print("Cannot divide by zero. Please provide a non-zero value for 'b'.")
        return None

num1 = 10
num2 = 0

result = divide_numbers(num1, num2)
if result is not None:
    print("Result:", result)
In this example, we have a function divide_numbers that takes two arguments and performs division. Inside the try block, we attempt to divide a by b. However, since dividing by zero is not allowed, it will raise a ZeroDivisionError if b is 0.

To handle this exception, we use the except ZeroDivisionError block. If a ZeroDivisionError occurs within the try block, Python will jump directly to the except block, execute the code inside it (in this case, printing an error message), and then continue the program flow after the try-except block.

When we run this code with num1 = 10 and num2 = 0, the output will be:

csharp
Copy code
Cannot divide by zero. Please provide a non-zero value for 'b'.
As you can see, the program didn't crash with an unhandled exception. Instead, it gracefully handled the ZeroDivisionError by displaying a user-friendly error message. The rest of the code after the try-except block is also executed without any issues.

Using try and except allows you to anticipate potential errors and deal with them gracefully, making your code more robust and preventing the program from crashing due to exceptions. It's a vital tool for writing reliable Python programs.
```

Q5. What are custom Exception in python? why do we need custom Exceptions? Explain with an example.

```python
Custom exceptions in Python are user-defined exceptions that extend the base Exception class or any other built-in exception class. They allow developers to define specific types of errors that are relevant to their application domain and provide meaningful error messages when exceptional situations occur.

We need custom exceptions in Python for the following reasons:

Clarity and Readability: Custom exceptions allow developers to create descriptive names for specific errors related to their application. This makes the code more readable and easier to understand for other developers working on the project.

Precise Error Handling: Custom exceptions help distinguish different types of errors and allow for precise error handling. This allows developers to catch and handle specific exceptions separately, providing appropriate actions or recovery strategies based on the nature of the exception.

Maintainability: By defining custom exceptions, you can centralize error handling logic related to your application's domain, making it easier to update and maintain the codebase.

Consistency: Custom exceptions allow you to enforce consistent error handling practices across your application, improving overall code quality and reliability.

Let's look at an example to illustrate the use of a custom exception:

python

class InsufficientFundsError(Exception):
    def __init__(self, balance, amount):
        self.balance = balance
        self.amount = amount

    def __str__(self):
        return f"Insufficient funds. Available balance: {self.balance}, Amount required: {self.amount}"


def withdraw_from_account(balance, amount):
    if amount > balance:
        raise InsufficientFundsError(balance, amount)
    return balance - amount


try:
    account_balance = 1000
    withdrawal_amount = 1500
    new_balance = withdraw_from_account(account_balance, withdrawal_amount)
    print("Withdrawal successful. New balance:", new_balance)
except InsufficientFundsError as e:
    print("Error:", e)
Output:

yaml

Error: Insufficient funds. Available balance: 1000, Amount required: 1500
In this example, we define a custom exception InsufficientFundsError that inherits from the base Exception class. This exception allows us to capture situations where a withdrawal amount exceeds the account balance.

In the withdraw_from_account function, we check if the withdrawal amount is greater than the account balance. If so, we raise the InsufficientFundsError with the available balance and the required amount as parameters.

In the try block, we attempt to withdraw an amount that exceeds the account balance. As a result, the InsufficientFundsError is raised and caught in the except block, where we display the custom error message indicating the available balance and the amount required for the withdrawal.

By using custom exceptions, we provide more context-specific error messages and handle the exceptional situation in a more controlled and meaningful way. This makes the code more robust and user-friendly, enhancing the overall quality of the application.
```

Q6. Create a custom exception class. Use this class to handle an exception.

```python
Sure! Let's create a custom exception class named InvalidInputError and use it to handle an exception in a simple program.

python
Copy code
class InvalidInputError(Exception):
    def __init__(self, message):
        self.message = message

    def __str__(self):
        return f"Invalid input: {self.message}"


def calculate_square_root(number):
    if number < 0:
        raise InvalidInputError("Cannot calculate the square root of a negative number.")
    return number ** 0.5


try:
    user_input = float(input("Enter a non-negative number: "))
    result = calculate_square_root(user_input)
    print(f"The square root of {user_input} is: {result}")
except InvalidInputError as e:
    print(f"Error: {e}")
In this example, we define a custom exception class InvalidInputError, which takes a message parameter and stores it as an instance variable.

The calculate_square_root function is designed to calculate the square root of a number. However, if the user enters a negative number, we raise the InvalidInputError with a custom message using the raise statement.

In the try block, we take user input and attempt to calculate the square root using the calculate_square_root function. If a negative number is provided, the InvalidInputError will be raised and caught in the except block, where we print the custom error message.

Now, let's see how the program works in different scenarios:

Valid input (e.g., 16):

yaml

Enter a non-negative number: 16
The square root of 16.0 is: 4.0
Invalid input (e.g., -5):

less

Enter a non-negative number: -5
Error: Invalid input: Cannot calculate the square root of a negative number.
Invalid input (e.g., "abc"):

vbnet

Enter a non-negative number: abc
Error: invalid input could not convert string to float: 'abc'
As you can see, the custom exception class InvalidInputError allows us to provide meaningful error messages when the user enters invalid input, making it easier to identify and handle specific exceptional situations.
```
